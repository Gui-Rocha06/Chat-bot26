```python
from groq import Groq # importando a biblioteca groq
```

```python
client = Groq(api_key=chat_api) # cria um cliente do Groq para realizar as requisições para a api
# Faz a solicitação para o modelo
completion =client.chat.completions.create(
    model="llama-3.3-70b-versatile",
    messages=[
        {
            "role":"system","content":"Você é um assistente de agricultor"
        },
        {
            "role":"user","content":"Quais sementes são boas \n para plantar no verão?"
        }
    ],
    temperature =0,
    max_tokens=1024,
    top_p=1,
    stream = False,

)

# Exibe a resposta gerada
print(" \n Resposta do assistente: \n")
print(completion.choices[0].message.content)
```

```python
client = Groq(api_key=chat_api) # cria um cliente do Groq para realizar as requisições para a api
while True:
  # Solicitando para que o usuário digite a pergunta
  user_prompt = input('\n Digite sua pergunta sobre agricultura ou sigite 15 para sair:')

  # Faz a solicitação para o modelo
  completion =client.chat.completions.create(
      model="llama-3.1-8b-instant",
      messages=[
          {
              "role":"system","content":"Você é um assistente de agricultor"
          },
          {
              "role":"user","content":user_prompt
          }
      ],
      temperature = 0.5,
      max_tokens=1024,
      top_p=1,
      stream = False,

)

  # Exibe a resposta gerada
  print(" \n Resposta do assistente: \n")
  print(completion.choices[0].message.content)

  if user_prompt.strip() == '15':
    print('Encerrando o chatbot')
    break
```

```python
client = Groq(api_key=chat_api) # cria um cliente do Groq para realizar as requisições para a api
while True:
  # Solicitando para que o usuário digite a pergunta
  user_prompt = input('\n Digite sua pergunta sobre agricultura ou digite 15 para sair:')
  if user_prompt.strip() == '15':
    print('Encerrando o chatbot')

    break
  else:
    # Faz a solicitação para o modelo
    completion =client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {
                "role":"system","content":"Você é um assistente de agricultor"
            },
            {
                "role":"user","content":user_prompt
            }
        ],
        temperature = 0.5,
        max_tokens=1024,
        top_p=1,
        stream = False,

)

  # Exibe a resposta gerada
  print(" \n Resposta do assistente: \n")
  print(completion.choices[0].message.content)

```
