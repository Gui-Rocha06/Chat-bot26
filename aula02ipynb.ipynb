```python
from google.colab import userdata
chat_api = userdata.get('api_chat')
```

```python
#chat_api
```

```python
from llama_index.llms.groq import Groq  # importando o metodo Groq a partir do framework llama
```

```python
llm = Groq(model='llama-3.3-70b-versatile',api_key=chat_api)
response = llm.complete('Qual é a substância do aroma do alecrim para uma criança de 10 anos')
```

```python
response
```

```python
paragrafos = response.text.split("\n\n")
for paragrafos in paragrafos:
  print(paragrafos)
print()
```
