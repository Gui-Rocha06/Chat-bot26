```python
import gradio as gr
from groq import Groq
```

```python
# Cria um cliente pra receber um modelo LLM
client = Groq(api_key = api_chat)

# Cria uma função pro assistente de quimica e fisica
def assistente_qui_fis(user_prompt) -> str:

  try:
    if user_prompt.strip()=="15":
      return "Encerrando o Chatbot"

    completion = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=[
            {"role":"system","content":"Voce é um assistente de química e física"},
            {"role":"user","content":user_prompt},
        ],
        temperature=0,
        max_tokens=1024,
        top_p=1,
        stream=False
    )

    # No Gradio devolva a resposta
    return completion.choices[0].message.content

  except Exception as e:
    print("\n Erro no assistente")
    return f"Erro: {e}"


# Cria a interface com o Gradio
iface = gr.Interface(
    fn=assistente_qui_fis,
    inputs= gr.Textbox(lines=10, label="Pergunte ao assistente sobre química ou física"),
    outputs= gr.Textbox(lines=10, label="Resposta"),
    title="Assistente de química e física",
    description="Faça perguntas sobre química e física e receba resposta de uma LLM especialista",
    live = False
)

if __name__=="__main__":
  iface.launch(debug=True,show_error=True)
```
