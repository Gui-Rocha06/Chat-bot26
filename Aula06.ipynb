```python
# importando a biblioteca
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
```

```python
# Criando classe personalizada para integrar ao Chroma 8
class ChromaEmbeddingWrapper:
  def __init__(self,model_name:str):  # inicializa com o modelo em embedding
    self.model = HuggingFaceEmbedding(model_name=model_name) # converte a entrada para o formato compativel com o Hugging face

  def __call__(self,input):
    return self.model.embed(input) # retorna o embedding da entrada

```

```python
embed_model_chroma = ChromaEmbeddingWrapper(model_name='intfloat/multilingual-e5-large')
```

```python
!pip install -q llama-index-vector-stores-chroma
```

```python
#Importando a chromaDB
import chromadb
db = chromadb.PersistentClient(path='./chroma_db')
chroma_client = db
collection_name = 'documentos_serenattto'

try:
  chroma_collection = chroma_client.get_or_create_collection(
      name=collection_name,
      embedding_function=embed_model_chroma
  )
except Exception as e:
  print(f'Erro ao criar coleçao: {e}')
```
